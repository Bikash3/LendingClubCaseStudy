```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
```

```python
# importing data from csv file
df = pd.read_csv('/content/drive/MyDrive/Upgrad/LendingClubCaseStudy/loan.csv')
df.shape
```

```python
# finding nul values for each columns
df.isnull().sum().sort_values(ascending=False)
```

```python
# removing columns only if all its values are null
df1 = df.dropna(axis=1, how='all')
df1.shape
```

```python
# checking percentage of null values in each columns and arrenging them in descending order
round((100*df1.isnull().sum())/len(df1.index)).sort_values(ascending=False)
```

```python
# removing coloums that have more then 30% of null values
columns_to_remove = ['next_pymnt_d', 'mths_since_last_record', 'mths_since_last_delinq', 'desc']
df1 = df1.drop(columns=columns_to_remove, axis=1)
df1.shape
```

```python
# checking datatype of all columns
df1.dtypes
```

```python
# fixing columns that have less then 30% of null valuesnull values.
# filling null values with NA in emp_title column
df1['emp_title'] = df1['emp_title'].fillna('NA')
```

```python
# filling null values with 0 in emp_length column
df1['emp_length'] = df1['emp_length'].fillna(0)
```

```python
# filling null values with 0 in pub_rec_bankruptcies column
df1['pub_rec_bankruptcies'] = df1['pub_rec_bankruptcies'].fillna(0)
```

```python
df1.head()
```

```python
# changing int_rate col to float datatype
df1['int_rate'] = df1['int_rate'].str.replace('%','').astype(float)
```

```python
df1.info()
```

```python
# take out month and year from issue_d into a new columns
df1['issue_month'] = pd.to_datetime(df1['issue_d'], format='%b-%y').dt.month
df1['issue_year'] = pd.to_datetime(df1['issue_d'], format='%b-%y').dt.year
```

```python
# selecting columns for the analysis
columns_to_keep = [
    'id',
    'loan_amnt',
    'verification_status',
    'home_ownership',
    'int_rate',
    'term',
    'purpose',
    'issue_year',
    'issue_month',
    'emp_length',
    'annual_inc',
    'dti',
    'grade',
    'installment',
    'addr_state',
    'delinq_2yrs',
    'loan_status'
                   ]
df_analysis = df1[columns_to_keep]
df_analysis = df_analysis[df_analysis['emp_length']!=0]
```

```python
df_analysis.head()
```

```python
loan_counts_by_year = df_analysis.groupby('issue_year')['id'].count().reset_index()
loan_counts_by_year.rename(columns={'id': 'loan_count'}, inplace=True)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(loan_counts_by_year['issue_year'], loan_counts_by_year['loan_count'])
plt.xlabel('Issue Year')
plt.ylabel('Number of Loans')
plt.title('Number of Loans Issued Each Year')
plt.xticks(loan_counts_by_year['issue_year'])  # Set x-axis ticks to be the issue years
plt.grid(axis='y')  # Add grid lines to the y-axis
plt.show()
```

# In the above univariate analysis, Substantial growth between 2007-2011 in the number of loan applicants can be observed.

```python
crg_off = pd.crosstab(df_analysis['issue_year'], df_analysis['loan_status'][df_analysis['loan_status']=='Charged Off'])
crg_off.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Charged off Loan Status Distribution by Issue Year')
plt.xlabel('Issue Year')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.legend(title='Loan Status')
plt.tight_layout()
plt.show()
```

# Substantial growth between 2007-2011 in the number of cherged off loans can be observed.

```python
# Group data by issue_year and calculate the mean of delinq_2yrs
delinq_by_year = df_analysis.groupby('issue_year')['delinq_2yrs'].mean().reset_index()
```

# Creates a cross-tabulation table followed by a stacked bar chart to visualize the distribution of loan statuses over different issue years.

```python
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.plot(delinq_by_year['issue_year'], delinq_by_year['delinq_2yrs'], marker='o')  # Use markers for better visualization
plt.title('Average Delinquencies (delinq_2yrs) Over Issue Years')
plt.xlabel('Issue Year')
plt.ylabel('Average Delinquencies (delinq_2yrs)')
plt.grid(True)  # Add grid lines for better readability
plt.show()
```

# Line plot reveals a general increase in delinquency rates over the years, which could indicate that the quality of the loans issued has deteriorated over time, resulting in more defaults. However, it is important to keep in mind that other factors, such as changes in the broader economic environment, variations.

```python
# Calculate the percentage of charged-off loans for each verification status
loan_status_by_verification = pd.crosstab(df_analysis['verification_status'], df_analysis['loan_status'])
charged_off_percentage=loan_status_by_verification['Charged Off']/loan_status_by_verification.sum(axis=1) * 100
```

```python
plt.figure(figsize=(10, 6))
plt.bar(charged_off_percentage.index, charged_off_percentage.values)
plt.title('Charged-Off Percentage by Verification Status')
plt.xlabel('Verification Status')
plt.ylabel('Charged-Off Percentage (%)')
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())  # Format y-axis as percentage
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.tight_layout()
plt.show()
```

# Most of the defaulters are coming under Verified Customers.

```python
verification_by_state = pd.crosstab(df_analysis['addr_state'], df_analysis['verification_status'])
verification_by_state_normalized = verification_by_state.div(verification_by_state.sum(axis=1), axis=0) * 100
```

```python
plt.figure(figsize=(12, 8))  # Adjust figure size as needed
sns.heatmap(verification_by_state_normalized, annot=True, cmap="YlGnBu", fmt=".1f")  # Use annot=True to display values
plt.title('Verification Status Distribution by State (Heatmap)')
plt.xlabel('Verification Status')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```
